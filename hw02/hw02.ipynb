```python
chronic['Question'].unique()
```

```python
crudePrevalence=chronic[chronic['DataValueType']=='Crude Prevalence']
question=crudePrevalence[(crudePrevalence['Question']=='Poverty')|(crudePrevalence['Question']=='Binge drinking prevalence among adults aged >= 18 years')]
df=question[(question['StratificationCategory1']=='Gender')|(question['StratificationCategory1']=='Overall')]
df.head()
df2=df.drop(df[(df['StratificationCategory1'] == 'Gender')&(df['Question'] == 'Poverty')].index)
df2.head()
```

## Q2

```python

df2Wide=df2.pivot(['LocationDesc','LocationAbbr','YearStart'],columns=['Question','Stratification1'],values='DataValue')
df2Wide=df2Wide.reset_index()
```

## Q5

```python
growthrate=df2Wide.dropna(subset=['binge_all'])
growthratebystate_first=growthrate.groupby('state').first().drop(columns=["poverty","binge_female","binge_male"])

growthratebystate_last=growthrate.groupby('state').last().drop(columns=["poverty","binge_female","binge_male"])

growthratebystate_first.head()
```

```python
growthratebystate_last.head()
```

```python
growthratebystate=pd.merge(growthratebystate_first,growthratebystate_last, on= ["state"], how="left")
growthratebystate['binge_all_y']=growthratebystate['binge_all_y'].astype('float')
growthratebystate['binge_all_x']=growthratebystate['binge_all_x'].astype('float')
growthratebystate['year_y']=growthratebystate['year_y'].astype('float')
growthratebystate['year_x']=growthratebystate['year_x'].astype('float')
growthratebystate['average annual growth']=(growthratebystate['binge_all_y']-growthratebystate['binge_all_x'])/(growthratebystate['year_y']-growthratebystate['year_x'])
growthratebystate.head()
```

```python
growthratebystate_top=growthratebystate.nlargest(5, columns="average annual growth")
growthratebystate_top
```

```python
growthratebystate_bottom=growthratebystate.nsmallest(5, columns="average annual growth")
growthratebystate_bottom
```
